```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
imgs_id = np.load('imgs_id_train.npy')
```

```python
'data/train/image/'+str(imgs_id[0]) + '.tif'
```

```python
i = 0
pred_dir = 'out'
for img_id in imgs_id:
#     print(img_id)
    img = cv2.imread(r'data/train/image/'+str(img_id) + '.tif')
    img_label = cv2.imread(r'data/train/label/'+str(img_id)+'_mask.tif') 
    img_pred = cv2.imread(r'preds/'+str(img_id)+'_pred.png')
    img_pred = cv2.resize(img_pred,(580,420), interpolation = cv2.INTER_CUBIC)
    img_label_gray = cv2.cvtColor(img_label,cv2.COLOR_BGR2GRAY)
    img_pred_gray = cv2.cvtColor(img_pred,cv2.COLOR_BGR2GRAY)
    ret_l,thresh_l = cv2.threshold(img_label_gray,127,255,0)
    ret_p,thresh_p = cv2.threshold(img_pred_gray,127,255,0)
    _l,contours_l, hierarchy_l = cv2.findContours(thresh_l,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    _p,contours_p, hierarchy_p = cv2.findContours(thresh_p,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(img, contours_l, -1, (0,255,0), 2)
    cv2.drawContours(img, contours_p, -1, (255,0,0), 2)
    if not os.path.exists(pred_dir):
        os.mkdir(pred_dir)
    cv2.imwrite('out/'+str(img_id)+'_out.png',img)
    if i % 100 == 0:
        print('Done: {0}/{1} images'.format(i, 2323))
    i += 1
```
